```python
%matplotlib inline
import pylab as plt
import matplotlib.patches

from descolors import BAND_COLORS
```

```python
plt.figure(figsize=(8,6))
for i,(n,v) in enumerate(BAND_COLORS.items()):
    print n,v
    r = matplotlib.patches.Rectangle((i,-i),6,1,color=v,label=n)
    plt.gca().add_artist(r)
    plt.plot(float('nan'),float('nan'),marker='s',ms=15,color=v,label=n)
    plt.annotate('%s-band'%n,((i-0.8),-i+0.3),xycoords='data',fontsize=10)
    plt.annotate(v,((i-0.8),-i),xycoords='data',fontsize=10)

plt.xlim(-1,len(BAND_COLORS))
plt.ylim(-len(BAND_COLORS),1)
plt.title("Standard DES Filter Colors")
legend = plt.legend(loc='lower left')
```
